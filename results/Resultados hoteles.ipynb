```python
import pandas as pd
import matplotlib.pyplot as plt

# Cargar los datos
df_alojamientos_paris = pd.read_csv('../data/alojamientosParís.csv')
df_alojamientos_roma = pd.read_csv('../data/alojamientosRoma.csv')
```

Big Numbers

```python
# 1. Distribución del Tipo de Alojamiento
tipo_alojamiento_paris = df_alojamientos_paris['accommodation_type_name'].value_counts()
tipo_alojamiento_paris
```

```python
tipo_alojamiento_roma = df_alojamientos_roma['accommodation_type_name'].value_counts()
print(tipo_alojamiento_roma)
```

```python
# 2. Puntuación de Reseñas Media

puntuacion_media_paris = df_alojamientos_paris['review_score'].mean()
print(puntuacion_media_paris)
```

```python
puntuacion_media_roma = df_alojamientos_roma['review_score'].mean()
print(puntuacion_media_roma)
```

```python
# 3. Precio Promedio por Noche

precio_promedio_noche_paris = df_alojamientos_paris['price_night'].mean()
print(precio_promedio_noche_paris)
```

```python
precio_promedio_noche_roma = df_alojamientos_roma['price_night'].mean()
print(precio_promedio_noche_roma)
```

```python
# 4. Distancia Promedio

distancia_promedio_paris = df_alojamientos_paris['distance'].mean()
print(distancia_promedio_paris)
```

```python
distancia_promedio_roma = df_alojamientos_roma['distance'].mean()
print(distancia_promedio_roma)
```

```python
# 5. Distribución de Clases de Alojamiento
distribucion_clases_paris = df_alojamientos_paris['class'].value_counts()
print(distribucion_clases_paris)
```

```python
distribucion_clases_roma = df_alojamientos_roma['class'].value_counts()
distribucion_clases_roma
```

Graficas

```python
plt.figure(figsize=(10, 6))

# Gráfico de Distribución del Tipo de Alojamiento
plt.subplot(2, 2, 1)
tipo_alojamiento_paris.plot(kind='bar', color='blue', alpha=0.7, label='París')
tipo_alojamiento_roma.plot(kind='bar', color='orange', alpha=0.7, label='Roma')
plt.title('Distribución del Tipo de Alojamiento')
plt.xlabel('Tipo de Alojamiento')
plt.ylabel('Cantidad')
plt.xticks(rotation=45)
plt.legend()

# Gráfico de Puntuaciones Medias
plt.subplot(2, 2, 2)
plt.bar(['París', 'Roma'], [puntuacion_media_paris, puntuacion_media_roma], color=['blue', 'orange'])
plt.title('Puntuación Media de Reseñas')
plt.ylabel('Puntuación')

# Gráfico de Precio Promedio por Noche
plt.subplot(2, 2, 3)
plt.bar(['París', 'Roma'], [precio_promedio_noche_paris, precio_promedio_noche_roma], color=['blue', 'orange'])
plt.title('Precio Promedio por Noche')
plt.ylabel('Precio ($)')

# Gráfico de Distancia Promedio
plt.subplot(2, 2, 4)
plt.bar(['París', 'Roma'], [distancia_promedio_paris, distancia_promedio_roma], color=['blue', 'orange'])
plt.title('Distancia Promedio a Puntos de Interés')
plt.ylabel('Distancia (km)')

# Ajustar el layout
plt.tight_layout()
plt.savefig('alojamientos_comparativo.png')  # Guardar el gráfico como imagen
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Cargar los datos
df_actividades_paris = pd.read_csv('../data/actividades_en_Paris.csv')
df_actividades_roma = pd.read_csv('../data/actividades_en_Roma.csv')



```

```python
def analizar_actividades_ciudad(df, nombre_ciudad):
    # 5 actividades más baratas
    actividades_mas_baratas = df.sort_values(by="Precio (Por persona)", ascending=True).head(5)

    # 5 actividades más caras
    actividades_mas_caras = df.sort_values(by="Precio (Por persona)", ascending=False).head(5)

    # 5 actividades con mejor valoración
    actividades_mejor_valoradas = df.sort_values(by="Valoracion", ascending=False).dropna().head(5)

    # 5 actividades con más opiniones
    actividades_mas_opiniones = df.sort_values(by="Nº Opiniones", ascending=False).dropna().head(5)

    return actividades_mas_baratas, actividades_mas_caras, actividades_mejor_valoradas, actividades_mas_opiniones
```

Big Numbers

```python
# Obtener los dataframes para París
actividades_mas_baratas_paris, actividades_mas_caras_paris, actividades_mejor_valoradas_paris, actividades_mas_opiniones_paris = analizar_actividades_ciudad(df_actividades_paris, "París")

# Obtener los dataframes para Roma
actividades_mas_baratas_roma, actividades_mas_caras_roma, actividades_mejor_valoradas_roma, actividades_mas_opiniones_roma = analizar_actividades_ciudad(df_actividades_roma, "Roma")
```

Graficas

```python
# Gráfico 1: Comparar la cantidad de actividades entre París y Roma
fig, ax1 = plt.subplots(figsize=(8, 6))
num_actividades_paris = df_actividades_paris.shape[0]
num_actividades_roma = df_actividades_roma.shape[0]
ciudades = ['París', 'Roma']
num_actividades = [num_actividades_paris, num_actividades_roma]

ax1.bar(ciudades, num_actividades, color=['blue', 'orange'])
ax1.set_title('Comparación de la cantidad de actividades entre París y Roma')
ax1.set_ylabel('Cantidad de actividades')

plt.show()
```

```python
# Gráfico 2: Comparar los precios promedio de las actividades entre París y Roma
# Calculamos el precio promedio de las actividades
precio_promedio_paris = df_actividades_paris['Precio (Por persona)'].mean()
precio_promedio_roma = df_actividades_roma['Precio (Por persona)'].mean()

fig, ax2 = plt.subplots(figsize=(8, 6))
precios_promedio = [precio_promedio_paris, precio_promedio_roma]

ax2.bar(ciudades, precios_promedio, color=['green', 'red'])
ax2.set_title('Comparación de los precios promedio de actividades entre París y Roma')
ax2.set_ylabel('Precio promedio (por persona en euros)')

plt.show()

```
